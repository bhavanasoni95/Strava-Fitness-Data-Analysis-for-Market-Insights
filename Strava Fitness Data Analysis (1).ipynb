# Types of Intrepretations


### Intrepretation 1 - User Activity Behavior: Steps per Day

```python
# Step category function
def step_category(steps):
    if steps < 3000:
        return 'Sedentary'
    elif steps < 7500:
        return 'Lightly Active'
    elif steps < 10000:
        return 'Active'
    else:
        return 'Very Active'
```

```python
# Apply to data
daily_activity['StepCategory'] = daily_activity['totalsteps'].apply(step_category)
```

```python
# Plot step category distribution
sns.countplot(data=daily_activity, x='StepCategory', order=['Sedentary', 'Lightly Active', 'Active', 'Very Active'], palette='pastel')
plt.title('User Distribution by Step Category')
plt.xlabel('Step Level')
plt.ylabel('Number of Days')
plt.show()
```

#### Interpretation:
- Most users do not meet 10,000 steps/day.
- The largest group of users falls into the "Very Active" category, followed closely by "Lightly Active," indicating a broad spectrum of user activity levels.
- Despite a significant number of "Very Active" users, there's also a substantial "Sedentary" group, suggesting an opportunity for targeted interventions to encourage more movement.

### Intrepretation 2 - Calorie Burn Trend vs Steps

```python
# Scatterplot: Steps vs Calories
sns.scatterplot(data=daily_activity, x='totalsteps', y='calories', hue='StepCategory', palette='muted')
plt.title('Calories Burned vs Total Steps')
plt.xlabel('Total Steps')
plt.ylabel('Calories')
plt.legend(title='Step Category')
plt.show()
```

#### Interpretation:
- More steps = more calories burned. Useful to promote step goals.
- The chart clearly illustrates the positive relationship between steps taken and calories burned across all activity categories, with "Very Active" users consistently burning the most calories at higher step counts.
- "Sedentary" and "Lightly Active" users generally fall into lower step and calorie ranges, highlighting an opportunity for Bellabeat to encourage increased activity within these segments to boost calorie expenditure.

### Intrepretation 6 - Sleep vs Steps Correlation

```python
daily_sleep.columns = daily_sleep.columns.str.lower()
daily_sleep.columns
```

```python
# Converting id type from int to string because they should be unique and can't any mathematical operation is applied
daily_sleep['id'] = daily_sleep['id'].astype(str)
```

```python
# Ensure date columns are in datetime format
daily_activity['activityDate'] = pd.to_datetime(daily_activity['activitydate'])
daily_sleep['sleepday'] = pd.to_datetime(daily_sleep['sleepday'])

# Merge sleep + activity on ID and Date
merged = pd.merge(daily_activity, daily_sleep, 
                  left_on=['id', 'activitydate'], 
                  right_on=['id', 'sleepday'], 
                  how='inner')

# Convert sleep to hours
merged['SleepHours'] = merged['totalminutesasleep'] / 60
```

```python
# Scatterplot
sns.scatterplot(data=merged, x='totalsteps', y='SleepHours', color='green')
plt.title('Relationship Between Daily Steps and Sleep Duration')
plt.xlabel('Total Steps')
plt.ylabel('Sleep Hours')
plt.show()
```

#### Interpretation:
- Users who walk more tend to sleep better — valuable for wellness campaigns.
